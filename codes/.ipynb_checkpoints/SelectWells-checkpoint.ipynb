```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as ply
import plotly.graph_objs as go
#import geopandas as gpd
```

```python
# reading a csv file containing well names and UTM coordenates, using pandas:
header = ['Well_Name', 'UTM_E', 'UTM_N']
data = pd.read_csv('../data/Coordenadas_MP.csv', sep=',', names=header, header=0)
data
```

```python
#creating a list of dictionary to save the values of each columns:
wells = []
for i in range(len(data)):
    dicio = {
        'wellname': data['Well_Name'][i],
        'utme': data['UTM_E'][i],
        'utmn': data['UTM_N'][i]
    }
    wells.append(dicio)
```

```python
plt.figure(figsize=(10,10))
for well in wells:
    plt.scatter(well['utme'], well['utmn'], color='grey')
    plt.annotate(well['wellname'], xy=(well['utme'],well['utmn']), xytext=[-10,10], textcoords='offset points')
plt.show()
```

```python
wells[50]
```

```python
xchoice = wells[50]['utme']
ychoice = wells[50]['utmn']
```

```python
dist = []
for well in wells:
    d = np.sqrt((xchoice - well['utme'])**2 + (ychoice - well['utmn'])**2)
    dist.append(d)
```

```python
numberofwells = 4
wellschosen = []
for i in np.argsort(dist)[:numberofwells]:
    wellschosen.append(wells[i])
```
